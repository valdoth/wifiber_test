```python
import pandas as pd
```

```python
price = pd.read_csv('../data/prices.csv', sep=";")
```

```python
# Séparer la colonne 'OFFRE,PRICE' en deux colonnes 'OFFRE' et 'PRICE'
price[['OFFRE', 'PRICE']] = price['OFFRE,PRICE'].str.split(',', expand=True).loc[:, [0, 1]]
```

```python
# Supprimer la colonne originale
df = price.drop(columns=['OFFRE,PRICE'])
```

```python
df_filtered = df_filtered.sort_values(by="PRICE")
```

```python
df_filtered = df_filtered.drop_duplicates(subset="OFFRE", keep="last")
```

```python
df_filtered['OFFRE'] = df_filtered['OFFRE'].str.replace('_', ' ', regex=False)
df_filtered['OFFRE'] = df_filtered['OFFRE'].str.replace('Bundle ', '', regex=False)
df_filtered['OFFRE'] = df_filtered['OFFRE'].str.strip()
```

```python
df_filtered["OFFRE"] = df_filtered["OFFRE"].str.lower()
```

```python
df_filtered.to_csv("../data/unique_price.csv", index=False)
```
