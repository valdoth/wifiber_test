```python
# Fonction pour convertir les valeurs en octets
def convert_to_bytes(value):
    if " Go" in value:
        number = float(value.split(' Go')[0])
        return number * (1024 ** 3)
    elif 'Go' in value:
        number = float(value.split('Go')[0])
        return number * (1024 ** 3)
    elif "VOIX" in value or "SMS" in value:
        return 0
    elif "Instagram" in value:
        return 0
    return value
```

```python
# Application de la conversion à la colonne 'QUANTITY'
df['QUANTITY'] = df['QUANTITY'].apply(convert_to_bytes)
```

```python
import pandas as pd

# Exemple de dataframe
monthly_explode = pd.DataFrame({
    'ID': [1, 2, 3],
    'OFFRE': [['A', 'B'], ['A', 'C'], ['B', 'C', 'D']]
})

# Exploser les listes d'offres dans des colonnes
offers_dummies = monthly_explode['OFFRE'].apply(pd.Series).stack()
offers_dummies = pd.get_dummies(offers_dummies, prefix='', prefix_sep='').groupby(level=0).sum()

# Joindre les colonnes dummies au dataframe original
monthly_explode = monthly_explode.drop(columns=['OFFRE']).join(offers_dummies)

print(monthly_explode)
```
